```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate sample dataset
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Function to initialize centroids
def initialize_centroids(X, k):
    indices = np.random.choice(len(X), k, replace=False)
    return X[indices]

# Function to assign clusters
def assign_clusters(X, centroids):
    distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
    return np.argmin(distances, axis=1)

# Function to update centroids
def update_centroids(X, labels, k):
    return np.array([X[labels == i].mean(axis=0) for i in range(k)])

# Task 1: Implement K-means Clustering from Scratch

# Main K-means function
def kmeans(X, k, iterations=100):
    centroids = initialize_centroids(X, k)
    for _ in range(iterations):
        labels = assign_clusters(X, centroids)
        centroids = update_centroids(X, labels, k)
    return centroids, labels

# Run k-means
centroids, labels = kmeans(X, 4)

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X')
plt.title("K-Means Clustering")
plt.show()
```

```python
# Task 2: Hierarchical Clustering with Dendrogram

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs

# Generate sample data
X, y = make_blobs(n_samples=200, centers=3, random_state=42)

# Create linkage matrix
linkage_matrix = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
#  Task 3: DBSCAN vs K-means on Noisy Data

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, KMeans

# Generate noisy moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
db_labels = dbscan.fit_predict(X)

# K-means
kmeans = KMeans(n_clusters=2, random_state=42)
km_labels = kmeans.fit_predict(X)

# Plot DBSCAN result
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=db_labels, cmap='plasma')
plt.title("DBSCAN Clustering")

# Plot K-means result
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=km_labels, cmap='viridis')
plt.title("K-means Clustering")
plt.show()
```

```python
 #  Task 4: Gaussian Mixture Model (GMM)

 from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Generate data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Fit GMM
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X)

# Plot GMM clusters
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='Accent')
plt.title("GMM Clustering")
plt.show()
```

```python
 # Task 5: Anomaly Detection (Isolation Forest and LOF)

 from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# Generate data with some noise
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=0.60, random_state=42)
X = np.vstack([X, np.random.uniform(low=-6, high=6, size=(20, 2))])  # Add anomalies

# Isolation Forest
iso = IsolationForest(contamination=0.05)
iso_preds = iso.fit_predict(X)

# Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
lof_preds = lof.fit_predict(X)

# Plot results
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=iso_preds, cmap='coolwarm')
plt.title("Isolation Forest")

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=lof_preds, cmap='coolwarm')
plt.title("Local Outlier Factor (LOF)")
plt.show()
```
